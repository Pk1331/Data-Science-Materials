```python
import numpy as np
```

## Linear Algebra Functions
- `np.dot(arr1, arr2)` - Dot product of two arrays.
- `np.matmul(arr1, arr2)` - Matrix product of two arrays.
- `np.cross(arr1, arr2)` - Cross product of two arrays.
- `np.linalg.inv(arr)` - Computes the inverse of a matrix.
- `np.linalg.eig(arr)` - Computes the eigenvalues and eigenvectors of a matrix.
- `np.linalg.det(arr)` - Computes the determinant of a matrix.
- `np.linalg.solve(a, b)` - Solves a linear matrix equation.

### `np.dot(arr1, arr2)` - Dot product of two arrays.
#### The **dot product** is the result of multiplying corresponding elements of two vectors and then summing them up.



```python
    arr1 = np.array([1, 2])
    arr2 = np.array([3, 4])
    result = np.dot(arr1, arr2)
    print(result)  # Output: 11
```

### `np.matmul(arr1, arr2)` - Matrix product of two arrays.


```python
arr1 = np.array([[1, 2], [3, 4]])
arr2 = np.array([[5, 6], [7, 8]])
result = np.matmul(arr1, arr2)
print(result)
```

### `np.cross(arr1, arr2)` - Cross product of two arrays.
#### The cross product is a way to multiply two vectors in 3D space, and the result is a new vector that is perpendicular to both of the original vectors.


```python
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])
result = np.cross(arr1, arr2)
print(result) 
```

### `np.linalg.det(arr)` - Computes the determinant of a matrix


```python
matrix = np.array([[1, 2], [3, 4]])
determinant = np.linalg.det(matrix)
print(determinant)
```

### `np.linalg.inv(arr)` - Computes the inverse of a matrix


```python
matrix = np.array([[1, 2], [3, 4]])
inverse = np.linalg.inv(matrix)
print(inverse)
```

### `np.linalg.eig(arr)` - Computes the eigenvalues and eigenvectors of a matrix.


```python
matrix = np.array([[1, 2], [3, 4]])
eigenvalues, eigenvectors = np.linalg.eig(matrix)
print(eigenvalues)
print(eigenvectors)
```

### `np.linalg.solve(a, b)` - Solves a linear matrix equation, or system of linear scalar equations.

```python
a = np.array([[3, 1], [1, 2]])
b = np.array([9, 8])
solution = np.linalg.solve(a, b)
print(solution)
```

### `np.linalg.svd(matrix)` - SVD is used in data compression and noise reduction.(**Singular Value Decomposition**)

```python
A= np.array([[3,2,2],[2,3,-2]])
U,S,V = np.linalg.svd(A)
print("U:", U)
print("S:", S)
print("V:", V)
```

##  Singular Value Decomposition (SVD)
### Example Matrix

```python
A= np.array([[3,2,2],[2,3,-2]])
print(A)
```

### Step 1: Compute A_T * A

```python
A_T = np.transpose(A)
print(A_T)
```

```python
A_T_A = np.matmul(A_T,A)
print(A_T_A)
```

### Step 2: Find Eigenvalues of A_T * A
#### The eigenvalues  satisfy the characteristic equation: **det(A<sup>T</sup> A - I) = 0**


```python
from sympy import symbols,Matrix,det
 = symbols('')
A_T_A = Matrix([
    [13 - , 12, 2],
    [12, 13 - , -2],
    [2, -2, 8 - ]
])
determinant = det(A_T_A)
determinant.simplify()
```

```python
from sympy import solve
quadratic_eq = -**2 + 34* - 225
eigenvalues = solve(quadratic_eq, )
eigenvalues
```

#### The eigenvalues are: <sub>1</sub> = 0, <sub>2</sub> = 9, <sub>3</sub> = 25


### Step 3: Compute Singular Values
- Singular values are always arranged in descending order of magnitude
#### <sub>1</sub> = 25 = 5, <sub>2</sub> = 9 = 3		



### Step 4: Construct the  Matrix
#### The  matrix is a diagonal matrix containing the singular values <sub>1</sub> and <sub>2</sub> Since  A is a 23 matrix,   will be a 23 matrix with the singular values on the diagonal:

```python
= np.array([[5,0,0],[0,3,0]])
print()
```

### Step 5: Compute Eigenvectors of A<sup>T</sup> A to Find V



```python
from sympy import symbols, Matrix, det
# Define A^T A matrix for the eigenvalue problem
A_T_A = Matrix([
    [13, 12, 2],
    [12, 13, -2],
    [2, -2, 8]
])

# Eigenvectors for 1 = 25, 2 = 9, 3 = 0
eigenvectors = A_T_A.eigenvects()
eigenvectors
```

#### These eigenvectors form the columns of the matrix V



```python
V = np.array([[1,1/4,-2],[1,-1/4,2],[0,1,1]])
print(V)
```

### Step 6: Compute the Matrix U
#### To find U, we calculate U = AV<sup>-1</sup>

```python
from sympy import Matrix
A = Matrix([
    [3, 2, 2],
    [2, 3, -2]
])

V = Matrix([
    [1, 1/4, -2],
    [1, -1/4, 2],
    [0, 1, 1]
])

Sigma_inv = Matrix([
    [1/5, 0, 0],
    [0, 1/3, 0]
])
U = A * V[:, :2] * Sigma_inv
U
```

## Random Numbers Functions
- `np.random.rand(d0, d1, ..., dn)` - Random values in [0,1).
- `np.random.randn(d0, d1, ..., dn)` - Random values from a normal distribution.
- `np.random.randint(low, high, size)` - Random integers from low to high.
- `np.random.choice(arr, size)` - Randomly select elements from an array.
- `np.random.shuffle(arr)` - Shuffles the elements in an array.
- `np.random.seed(seed)` - Seeds the random number generator.

### `np.random.rand(d0, d1, ..., dn)` - Random values in [0,1).

```python
random_values = np.random.rand(3)
print(random_values)
```

```python
random_values = np.random.rand(3, 2)
print(random_values)
```

### `np.random.randn(d0, d1, ..., dn)` - Random values from a normal distribution.

```python
random_normal = np.random.randn(2, 2)
print(random_normal)
```

### `np.random.randint(low, high, size)` - Random integers from low to high.

```python
random_integers = np.random.randint(10, 50, size=(3, 3))
print(random_integers)
```

### `np.random.choice(arr, size)` - Randomly select elements from an array.

```python
arr = np.array([1, 2, 3, 4, 5])
random_choice = np.random.choice(arr, size=3)
print(random_choice)
```

### `np.random.shuffle(arr)` - Shuffles the elements in an array.

```python
arr = np.array([1, 2, 3, 4, 5])
np.random.shuffle(arr)
print(arr)
```

### `np.random.seed(seed)` - Seeds the random number generator.
#### If you run this code again with the same seed (42), you will get exactly the same random numbers.

```python
np.random.seed(42)
random_values_seeded = np.random.rand(3, 2)
print(random_values_seeded)
```
