### **<span style="color:green">Filling null values</span>**

#### **df.fillna(value)**: Replace null values with a specific value.
**`df.fillna(value=None, method=None, axis=None, inplace=False, limit=None)`**

**`df["ColumnName"] = df["ColumnName"].fillna("ValueToReplace")`**


```python
#Before Filling Null Values Count
df2["country"].isnull().sum()
```

```python
df["country"] = df["country"].fillna("Unknown")
```

```python
#After Filling Null Values Count
df["country"].isnull().sum()
```

#### **1. Fill all NaNs with a single value**

```python
df2.fillna(0)
```

```python
df2.isnull().sum()
```

#### **2. Fill NaNs in a specific column**

```python
df2["country"].fillna("Unknown")
```

#### **3. Fill using a dictionary (column-wise)**

```python
df.fillna({'col-1': value1, 'col-2': value2})
```

#### **4. Forward fill (fill with previous row value):**

```python
df.isnull().sum()
```

#### **deprecated**

```python
df['agent'] = df['agent'].fillna(method="ffill") # deprecated
```

#### **new Approach**

```python
# new way
df['agent'] = df['agent'].ffill()
```

```python
df.isnull().sum()
```

#### **5. Backward fill (fill with next row value)**

```python
df['agent'] = df['agent'].fillna(method="bfill") # deprecated
```

```python
df['agent'] = df['agent'].bfill()
```

```python
df.isnull().sum()
```
