### **<span style="color:green">Drop null values</span>**

#### **df.dropna()**: Drop rows with null values.
#### **`df.dropna(axis=0, how='any', thresh=None, subset=None, inplace=False)`**
---

### **Parameters :**

| Parameter   | Description |
|-------------|-------------|
| `axis`      | `0` = drop rows (default), `1` = drop columns |
| `how`       | `'any'` = drop if **any** value is NaN, `'all'` = drop if **all** are NaN |
| `thresh`    | Keep row/column only if it has at least that many **non-NaN** values |
| `subset`    | Specify column(s) to look for NaNs |
| `inplace`   | `True` to change original DataFrame |

---


````python
df_copy = df.copy()
df_copy
````

````python
df_copy.isnull().sum()
````

````python
df_copy.shape
````

#### **1. Drop all rows with any NaN**

````python
df_copy.dropna()
````

#### **2. Drop all columns with any NaN**

````python
df_copy.dropna(axis=1)
````

#### **3. Drop rows where all values in that row are NaN**

````python
df_copy.dropna(how="all")
````

#### **4. Drop rows only if specific column has NaN**

````python
df_copy.dropna(subset=['country'])
````

#### **5. Drop rows with less than 2 non-NaN values**

````python
df_copy.dropna(thresh=2)
````

````python
df_copy.shape
````

#### **6. Drop rows inplace (modifies df)**

````python
df_copy.dropna(inplace=True)
````

````python
df_copy.shape
````

## **<center><span style="color:brown">Handling Duplicates</span></center>**

### **<span style="color:green">Check Duplicate values</span>**

#### **df.duplicated()** : Find duplicate rows in a DataFrame
#### **`df.duplicated(subset=None, keep='first')`**

---

### 🔍 **Parameters:**

| Parameter | Description |
|----------|-------------|
| `subset` | Column(s) to check for duplicates (default: all columns) |
| `keep`   | `'first'` (default), `'last'`, or `False` <br> `'first'`: keeps first duplicate as `False`, rest `True` <br> `'last'`: keeps last duplicate as `False` <br> `False`: marks all duplicates as `True` |

---



````python
df1 = pd.read_csv("datasets/duplicated_dataset.csv")
df1
````

````python
df1.duplicated()
````

#### **Filter duplicate rows**

````python
df1[df1.duplicated()]
````

### **<span style="color:green">Drop Duplicate values</span>**

#### **drop_duplicates()**: removes duplicate rows from a DataFrame


````python
df1_copy = df1.copy()
````

#### **1. Remove all duplicated rows (keep first):**

````python
df1_copy = df1_copy.drop_duplicates()
df1_copy
````

#### **2. Drop duplicates based on specific columns**

````python
df1 = df1.drop_duplicates(subset=['Name', 'Age'])
df1
````

#### **3. Keep last occurrence instead of first**

````python
df1 = df1.drop_duplicates(keep='last')
df1
````

#### **4. Remove all duplicates (no keeping)**

````python
df1 = df1.drop_duplicates(keep=False)
df1
````

#### **5. In-place removal**

````python
df1.drop_duplicates(inplace=True)
df1
````

````python
df1.loc[df1['Name'].str.startswith('A')]
````

````python
df1['Name'] = df1['Name'].replace(r'^A.*', 'NameStartsWithA', regex=True)
df1
````

````python
df1.loc[df1['Name'].str.startswith('Name')]
````

## **<center><span style="color:brown">Renaming Columns</span></center>**

#### **rename()** : Change column names or index labels
```python
df.rename(
    mapper=None,
    index=None,
    columns=None,
    axis=None,
    copy=True,
    inplace=False,
    level=None,
    errors='ignore'
)
```

---

### **Parameters**:

| Parameter   | Description |
|-------------|-------------|
| `mapper`    | Dict or function to rename labels |
| `index`     | Dict for renaming row index |
| `columns`   | Dict for renaming column names |
| `axis`      | 0 or ‘index’, 1 or ‘columns’ |
| `copy`      | Default `True`. Copy data or not |
| `inplace`   | `True` to rename in same DataFrame |
| `level`     | If MultiIndex, level to apply |
| `errors`    | `ignore` (default) or `raise` for missing labels |

---


````python
df1 = pd.read_csv("datasets/duplicated_dataset.csv")
````

#### **1. Rename a single column**

````python
df1 = df1.rename(columns={'Name': 'FullName'})
df1
````

#### **2. Rename multiple columns**

````python
df1 = df1.rename(columns={'Name':'FullName','Age':'Years'})
df1
````

#### **3. Rename an index value**

````python
for i in df1.index:
    df1 = df1.rename(index={i:f'row-{i}'})
df1
````

#### **4. Rename in-place (without creating new DataFrame)**

````python
df1.rename(columns={'Score': 'Marks'}, inplace=True)
df1
````

## **<center><span style="color:brown">Changing Data Types</span></center>**

#### **astype()** : Convert the data type of a column

```python
DataFrame.astype(
    dtype,
    copy=True,
    errors='raise'
)
```

---

### **Parameters :**

| Parameter | Description |
|-----------|-------------|
| `dtype`   | The type(s) to cast one or more columns to (e.g., `'int'`, `'float'`, `'str'`, or a dict for multiple columns) |
| `copy`    | `True` (default): returns a copy <br>`False`: try to do it without copying |
| `errors`  | `'raise'` (default): throws error on failure <br>`'ignore'`: skips failed conversions |
`
---
### **Basic Syntax**:
**`df['column_name'] = df['column_name'].astype(new_type)`**


````python
df1 = pd.read_csv("datasets/duplicated_dataset.csv")
df1
````

````python
df1.dtypes
````

#### **1. Convert column to Str**

````python
df1['Age'] = df1['Age'].astype(str)
df1.dtypes
````

#### **2. Convert column to integer**

````python
df1['Age'] = df1['Age'].astype(int)
df1.dtypes
````

#### **3. Convert multiple columns at once**

````python
df1 = df1.astype({'Age':int , 'Score':float})
df1.dtypes
````

#### **4. Convert to category (for memory efficiency)**

````python
df1['Grade'] = df1['Grade'].astype('category')
df1.dtypes
````

## **Why does `category` improve performance in pandas?**

When you use `category` dtype instead of regular `object` (like strings), **pandas stores data more efficiently** — both in **memory** and **speed**. Here's why:

###  1. **Less Memory Usage**  
Instead of storing full strings again and again, pandas stores:
- a **unique list of values** once, and  
- then just stores **integer codes** for each row

**Example:**
```python
['A', 'B', 'A', 'B', 'A'] → ['A', 'B'] + [0, 1, 0, 1, 0]
```

This saves tons of memory if the column has **repeating values** (like Gender, Country, Grade, etc.)


### 2. **Faster Operations**  
Since it's working with **codes (integers)** instead of strings, operations like:
- filtering
- grouping
- sorting

become much faster.


### 3. **Good for Categorical Data**  
If your column contains fixed categories like:
- `'Male'`, `'Female'`
- `'Yes'`, `'No'`
- `'A'`, `'B'`, `'C'`

then `category` is **perfect**.


### Example:
```python
df['Gender'] = df['Gender'].astype('category')
```

## **<center><span style="color:brown">Filtering Data</span></center>**

#### **1. Filter by a Single Condition**
**Syntax**:
```python
    df[df['Column'] > value]
```


````python
df1[df1['Score'] > 95]
````

#### **2. Filter with Multiple Conditions**
**Syntax**:
```python
    df[(df['Col1'] > value1) & (df['Col2'] == 'something')]
    df[(df['Col1'] < value1) | (df['Col2'] != 'other')]

```

````python
df1[(df1['Score'] > 95 ) & (df1['Dept'] == "HR")]
````

#### **3. Filter String Values**

````python
df1[df1["Name"] == "Eva"]
````

#### **3.Filter Null / Not Null**

````python
df[df['country'].isnull()]
````

````python
df[df['country'].notnull()]
````
